```python
import pandas as pd
```

```python
df = pd.read_csv("PremierLeague.csv")
```

```python
df_final_copy = df[["MatchID","Season","MatchWeek","Date","HomeTeam","AwayTeam",
    "FullTimeHomeTeamGoals","FullTimeAwayTeamGoals","FullTimeResult",
    "HalfTimeHomeTeamGoals","HalfTimeAwayTeamGoals","HalfTimeResult",
    "HomeTeamShots","AwayTeamShots","HomeTeamShotsOnTarget","AwayTeamShotsOnTarget",
    "HomeTeamCorners","AwayTeamCorners","HomeTeamYellowCards","AwayTeamYellowCards",
    "HomeTeamRedCards","AwayTeamRedCards","HomeTeamPoints","AwayTeamPoints"]]
```

```python
df_final_copy["Date"] = pd.to_datetime(df_final_copy["Date"], errors= 'coerce', dayfirst=True)
```

```python
df_final_copy = df_final_copy[df_final_copy["Date"] > "2000-07-01"]
```

```python
df_final = df_final_copy
df_final
```

```python
df_final['Team_Attack_Strength'] = df_final.groupby('HomeTeam')['FullTimeHomeTeamGoals'].transform(lambda x: x.rolling(5, min_periods=1).mean())
df_final['Team_Defense_Strength'] = df_final.groupby('AwayTeam')['FullTimeAwayTeamGoals'].transform(lambda x: x.rolling(5, min_periods=1).mean())
df_final['Rolling_Goals_Scored'] = df_final.groupby('HomeTeam')['FullTimeHomeTeamGoals'].transform(lambda x: x.rolling(10, min_periods=1).mean())
df_final['Rolling_Goals_Conceded'] = df_final.groupby('AwayTeam')['FullTimeAwayTeamGoals'].transform(lambda x: x.rolling(10, min_periods=1).mean())

```

```python


df_final['Home_Win_Percentage'] = df_final.groupby('HomeTeam')['FullTimeResult'].transform(lambda x: x.eq('H').astype(float).rolling(10, min_periods=1).mean())
df_final['Away_Win_Percentage'] = df_final.groupby('AwayTeam')['FullTimeResult'].transform(lambda x: x.eq('A').astype(float).rolling(10, min_periods=1).mean())

df_final['H2H_Wins'] = df_final.groupby(['HomeTeam', 'AwayTeam'])['FullTimeResult'].transform(lambda x: x.eq('H').astype(float).rolling(5, min_periods=1).sum())
df_final['H2H_Draws'] = df_final.groupby(['HomeTeam', 'AwayTeam'])['FullTimeResult'].transform(lambda x: x.eq('D').astype(float).rolling(5, min_periods=1).sum())
df_final['H2H_Losses'] = df_final.groupby(['HomeTeam', 'AwayTeam'])['FullTimeResult'].transform(lambda x: x.eq('A').astype(float).rolling(5, min_periods=1).sum())

df_final['Home_Goals_Avg'] = df_final.groupby('HomeTeam')['FullTimeHomeTeamGoals'].transform(lambda x: x.rolling(5, min_periods=1).mean())
df_final['Away_Goals_Avg'] = df_final.groupby('AwayTeam')['FullTimeAwayTeamGoals'].transform(lambda x: x.rolling(5, min_periods=1).mean())

df_final['Home_ShotsOnTarget_Avg'] = df_final.groupby('HomeTeam')['HomeTeamShotsOnTarget'].transform(lambda x: x.rolling(5, min_periods=1).mean())
df_final['Away_ShotsOnTarget_Avg'] = df_final.groupby('AwayTeam')['AwayTeamShotsOnTarget'].transform(lambda x: x.rolling(5, min_periods=1).mean())

df_final['Home_RedCards_Avg'] = df_final.groupby('HomeTeam')['HomeTeamRedCards'].transform(lambda x: x.rolling(5, min_periods=1).mean())
df_final['Away_RedCards_Avg'] = df_final.groupby('AwayTeam')['AwayTeamRedCards'].transform(lambda x: x.rolling(5, min_periods=1).mean())
```

```python
import numpy as np
import statsmodels.api as sm

features = [
    'Home_Win_Percentage', 'Away_Win_Percentage', 
    'Home_Goals_Avg', 'Away_Goals_Avg', 
    'H2H_Wins', 'H2H_Losses', 'H2H_Draws', 
    'Home_ShotsOnTarget_Avg', 'Away_ShotsOnTarget_Avg', 
    'Home_RedCards_Avg', 'Away_RedCards_Avg'
]

df_final_copy = df_final.dropna(subset=features + ['FullTimeHomeTeamGoals', 'FullTimeAwayTeamGoals'])

y_home = df_final_copy['FullTimeHomeTeamGoals']
y_away = df_final_copy['FullTimeAwayTeamGoals']

X = sm.add_constant(df_final_copy[features])

home_goal_model = sm.GLM(y_home, X, family=sm.families.Poisson()).fit()
away_goal_model = sm.GLM(y_away, X, family=sm.families.Poisson()).fit()

print(home_goal_model.summary())
print(away_goal_model.summary())
```

```python
df_final_copy['Expected_Home_Goals'] = home_goal_model.predict(X)
df_final_copy['Expected_Away_Goals'] = away_goal_model.predict(X)
```

```python
import numpy as np
import pandas as pd

def monte_carlo_match_simulation(home_goals_avg, away_goals_avg, simulations=10000):
    """
    Simulates a match 10,000 times using Poisson-distributed goals.
    Returns estimated probabilities for Home Win, Draw, and Away Win.
    """
    home_goals = np.random.poisson(home_goals_avg, simulations)
    away_goals = np.random.poisson(away_goals_avg, simulations)
    
    home_wins = np.sum(home_goals > away_goals) / simulations
    draws = np.sum(home_goals == away_goals) / simulations
    away_wins = np.sum(home_goals < away_goals) / simulations
    
    return {
        'Home Win Probability': home_wins,
        'Draw Probability': draws,
        'Away Win Probability': away_wins
    }


match_data = df_final_copy[df_final_copy['MatchID'] == "2024-2025_West Ham_Leicester"].iloc[0]

home_goals_pred = match_data['Expected_Home_Goals']
away_goals_pred = match_data['Expected_Away_Goals']

probabilities = monte_carlo_match_simulation(home_goals_pred, away_goals_pred)

print(probabilities)

bookmaker_odds = {'Home Win': 2.00, 'Draw': 3.50, 'Away Win': 3.75}  

expected_values = {
    outcome: (probabilities[outcome + ' Probability'] * bookmaker_odds[outcome]) - 1
    for outcome in ['Home Win', 'Draw', 'Away Win']
}

print("Expected Values:", expected_values)
```
